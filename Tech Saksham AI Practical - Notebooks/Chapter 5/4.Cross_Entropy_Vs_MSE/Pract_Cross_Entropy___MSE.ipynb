#**MSE (Mean Squared Error)**
##Squared Error loss for each training example, also known as L2 Loss, is the square of the difference between the actual and the predicted values:
##L = ( y - f(x) )^2
##The corresponding cost function is the Mean of these Squared Errors (MSE).

Reference -1. https://www.analyticsvidhya.com/blog/2019/08/detailed-guide-7-loss-functions-machine-learning-python-code/

2.https://www.geeksforgeeks.org/python-mean-squared-error/

```python
from sklearn.metrics import mean_squared_error

# Given values
Y_true = [1,1,2,2,4] # Y_true = Y (original values)

# calculated values
Y_pred = [0.6,1.29,1.99,2.69,3.4] # Y_pred = Y'

# Calculation of Mean Squared Error (MSE)
mean_squared_error(Y_true,Y_pred)
```

#**Cross Entropy Loss**

##Also known as Negative Log Likelihood. It is the commonly used loss function for classification. Cross-entropy loss progress as the predicted probability diverges from actual label.

g1-21.png

Reference -
1.https://www.geeksforgeeks.org/ml-common-loss-functions/
2.https://machinelearningmastery.com/loss-and-loss-functions-for-training-deep-learning-neural-networks/



```python
from math import log

def binary_cross_entropy(actual, predicted):
	sum_score = 0.0
	for i in range(len(actual)):
		sum_score += actual[i] * log(1e-15 + predicted[i])
	mean_sum_score = 1.0 / len(actual) * sum_score
	return -mean_sum_score

Y_pred = [0,1,1,1,0]
Y_true = [1,1,1,1,1]

loss = binary_cross_entropy(Y_true,Y_pred)
print(loss)
```
